```python
# clip_box example
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms

fig, ax = plt.subplots(subplot_kw=dict(aspect='equal'))
ax.set_title('default clip box')
el = Ellipse((0, 0), 10, 20, facecolor='blue', alpha=0.5)
ax.add_artist(el)
ax.set(xlim=[-5, 5], ylim=[-5, 5])

fig, ax = plt.subplots(subplot_kw=dict(aspect='equal'))
ax.set_title('clip box out of axis bounds')
el = Ellipse((0, 0), 10, 20, facecolor='red', alpha=0.5)
ax.add_artist(el)
ax.set(xlim=[-5, 5], ylim=[-5, 5])
bbox = matplotlib.transforms.TransformedBbox(matplotlib.transforms.Bbox([[0, 1.2], [1, 0]]), ax.transAxes)
el.set_clip_box(bbox)
```

```python
# clip_box example
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms

fig, ax = plt.subplots(subplot_kw=dict(aspect='equal'))
ax.set_title('default clip True')
el = Ellipse((0, 0), 10, 20, facecolor='blue', alpha=0.5)
ax.add_artist(el)
ax.set(xlim=[-5, 5], ylim=[-5, 5])

fig, ax = plt.subplots(subplot_kw=dict(aspect='equal'))
ax.set_title('clip False')
el = Ellipse((0, 0), 10, 20, facecolor='red', alpha=0.5)
ax.add_artist(el)
ax.set(xlim=[-5, 5], ylim=[-5, 5])
el.set_clip_on(False)
```
